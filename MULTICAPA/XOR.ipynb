```python
import numpy as np

# Datos XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]]) ## entrada
y = np.array([[0],[1],[1],[0]])## salida
## importamos la libreria con los datos de entrada y salida
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

modelo = Sequential([
    Dense(4, activation='relu', input_dim=2),
    Dense(1, activation='sigmoid')
])
## nos ayuda a crear la red neuronal con las capas ocultas nos indica que es un ejrcicio no-lineal con relu
```

```python
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje deseada
learning_rate = 0.01
adam_optimizer = Adam(learning_rate=learning_rate)
## almacena los datos necesarios para el proceso de aprendizaje
```

```python
modelo.compile(
    optimizer=adam_optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
## le indicaremos como debe ser entrenado es decir adam ayuda a ajustar los pesos con la tasa de aprendizaje y cuantas veces predijo bien
```

```python
history = modelo.fit(
    X, y,
    epochs=50,
    verbose=1,
)
## carga el # de epocas para ver la prediciion en este caso sera de 50 veces entonces hara las operaciones necesarias
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.title('Evolución de la pérdida (Loss)')
plt.xlabel('Época')
plt.ylabel('Binary Cross-Entropy')
plt.show() ## nos muestra graficamente si funciona o no
```

```python

loss, accuracy = modelo.evaluate(X, y)
print(f'Loss: {loss}, Accuracy: {accuracy}')

## evaluamos si el modelo funciono 0 no osea si aprendio la logica del modelo
```

```python
pred = (modelo.predict(X) > 0.5).astype(int)
for i, (inp, p) in enumerate(zip(X, pred)):
    print(f"{inp} -> {p[0]}")

    ## con esto ya podemos observar los datos de salida y observar mejor si funciono
```
