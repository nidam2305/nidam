```python
class PRECEPTRON:
  def __init__(self,pesos,sesgo,tasa_de_aprendizaje):
     self.pesos=pesos
     self.sesgo=sesgo
     self.tasa_de_aprendizaje=tasa_de_aprendizaje
    # agregamos los valores iniciales usando a init como el constructoru

  def funcion_activacion(self,Z):
      if Z>=0:
        return 1
      else:
        return 0
          # saldra el resultado de la suma ponderada es decir el nuevo valor ahora llamado Z
  def ENTRENAMIENTO(self,X,Y,epocas):
      for epoca in range(epocas):
          for dato in X:
              X1,X2=dato[0],dato[1]
              W1=self.pesos[0]
              W2=self.pesos[1]
              B=self.sesgo
              Z=X1*W1+X2*W2+B
              Y_pred=self.funcion_activacion(Z)
              indice=X.index(dato)
              Y_real=Y[indice]
              self.pesos[0]=W1+self.tasa_de_aprendizaje*(Y_real-Y_pred)*X1
              self.pesos[1]=W2+self.tasa_de_aprendizaje*(Y_real-Y_pred)*X2
              self.sesgo=B+self.tasa_de_aprendizaje+(Y_real-Y_pred)
              print(f"Dato{dato},prediccion{Y_pred},pesos{self.pesos},sesgo{self.sesgo}")
              print(f"Fin de la epoca{epoca}")
                # esta parte toma todos los datos , hace la suma ponderada y si se equivoca lo ajusta
                #las espocas son las variables de X que son el conjunto de datos
                # el for que hace el bucle con las epocas que digamos
                #con el indice es la respuesta correcta que deberia de dar la sumna ponderada
```

```python
X=[[1, 1],[1, 0],[0, 1],[0, 0]]
Y=[1,0,0,0]
pesos=[2,3]
sesgo=4
tasa_de_aprendizaje=0.1
# se le otorgan los valores iniciales y aqui comienza a procesar la informacion dada
```

```python
p=PRECEPTRON(pesos,sesgo,tasa_de_aprendizaje)
# aqui ya sabe cuales son los valores que debe de usar
```

```python
p.ENTRENAMIENTO(X,Y,20)
# aqui podemos identifica que tenemos los datos de entrada las X y las epocas  el num 20
# aqui empieza el ciclo de la prueba donde ajustara los pesos y sesgos en cada columna
```
